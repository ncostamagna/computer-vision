# Image Segementation and the Watershed Algorithm

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap=cmap)
```

## Our Task: Draw Contours Around the Coins

-----------------

## Common Coin Example

## Naive Approach

Let's try to simply use a threshold and then use findContours.

### Example Image

```python
sep_coins = cv2.imread('../DATA/pennies.jpg')
```

##### Para un humano es facil decir que son monedas por separado, pero para muchos algoritmos de vision computacional es dificil distinguirlo

```python
display(sep_coins)
```

### Apply Median Blurring

We have too much detail in this image, including light, the face edges on the coins, and too much detail in the background. Let's use Median Blur Filtering to blur the image a bit, which will be useful later on when we threshold.

```python
# Aplicamos enfoque medio, nos deshacemos de las caracteristicas que no necesitamos
sep_blur = cv2.medianBlur(sep_coins,25)
```

```python
display(sep_blur)
```

```python
# Pasamos a gris
gray_sep_coins = cv2.cvtColor(sep_blur,cv2.COLOR_BGR2GRAY)
```

```python
display(gray_sep_coins,cmap='gray')
```

## Binary Threshold

```python
# Separacion binaria entre el fondo y el contorno
ret, sep_thresh = cv2.threshold(gray_sep_coins,160,255,cv2.THRESH_BINARY_INV)
```

```python
display(sep_thresh,cmap='gray')
```

## FindContours

```python
# BUscamos contornos
image, contours, hierarchy = cv2.findContours(sep_thresh.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
# For every entry in contours
for i in range(len(contours)):
    
    # last column in the array is -1 if an external contour (no contours inside of it)
    # Si es el contorno externo
    if hierarchy[0][i][3] == -1:
        # We can now draw the external contours from the list of contours
        cv2.drawContours(sep_coins, contours, i, (255, 0, 0), 10)
```

```python
display(sep_coins)
```

# Watershed Algorithm

-----------

Let's now try the watershed algorithm apporach to draw contours around the pennies. Also make sure to watch the video to understand what the function calls are doing here.

## Using the WaterShed Algorithm

#### Step 1: Read Image

```python
img = cv2.imread('../DATA/pennies.jpg')
```

#### Step 2: Apply Blur

```python
img = cv2.medianBlur(img,35)
```

#### Step 3: Convert to Grayscale

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

#### Step 4: Apply Threshold (Inverse Binary with OTSU as well)

```python
# Aplicamos OTSu metodo y luego invertimos los colores para que nos quede las monedas blancas
# https://en.wikipedia.org/wiki/Otsu%27s_method
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
display(thresh,cmap='gray')
```

### Optional Step 5: Noise Removal

```python
# noise removal
# Se remueve el ruido, es opcional y no hara la diferencia en modenas de centavos pero es un paso comun a usar
# cuando se trata de imagenes mas complejas
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2) # reucimos el ruido
```

```python
display(opening,cmap='gray')
# todavia seguimos teniendo el problema que las monedas se tocan
```

#### Step 6: Grab Background that you are sure of

```python
# sure background area
# dilatamos hasta que se toquen, nos aseguramos el fondo del area
# expande primer plano hasta que toque algun limite
sure_bg = cv2.dilate(opening,kernel,iterations=3)
```

```python
display(sure_bg,cmap='gray')
```

#### Step 7: Find Sure Foreground

```python
# Finding sure foreground area
# a medida que los ceros estan mas lejos de los pixels, va aumentando el valor
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)

# 0.7*dist_transform.max() -> es el limite, un 70% del valor maximo, con esto me aseguro de tomar los valores
# del medio
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)
```

```python
display(dist_transform,cmap='gray')
```

```python
display(sure_fg,cmap='gray')
```

#### Step 8: Find Unknown Region

```python
# Finding unknown region
sure_fg = np.uint8(sure_fg)
# sector desconocido, el resto de la monera que no sale en la imagen anterior,
# resto la imagen binaria de la moneda con los puntos de la imagen anterior
unknown = cv2.subtract(sure_bg,sure_fg)
```

```python
# No estoy seguro si el sector blanco de la imagen es parte de la moneda o el fondo
# tengo que pasarle al algoritmo las semillas de la imagen sure_fg para que pueda encontrar los segmentos
display(unknown,cmap='gray')
```

#### Step 9: Label Markers of Sure Foreground

```python
# Marker labelling
# conectamos los componentes e identificamos cada moneda
ret, markers = cv2.connectedComponents(sure_fg)
# Add one to all labels so that sure background is not 0, but 1
#Incremento en 1 para luego poder colocar en negro el sector desconocido de las monedas
markers = markers+1
# Now, mark the region of unknown with zero
markers[unknown==255] = 0
```

```python
display(markers,cmap='gray')
```

#### Step 10: Apply Watershed Algorithm to find Markers

```python
#pasamos los carcadores que definimos y la imagen original
markers = cv2.watershed(img,markers)
```

```python
display(markers)
```

#### Step 11: Find Contours on Markers

```python
# Ahora volvemos a buscar los contornos con los marcadores que generamos
image, contours, hierarchy = cv2.findContours(markers.copy(), cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)

# For every entry in contours
for i in range(len(contours)):
    
    # last column in the array is -1 if an external contour (no contours inside of it)
    if hierarchy[0][i][3] == -1:
        
        # We can now draw the external contours from the list of contours
        cv2.drawContours(sep_coins, contours, i, (255, 0, 0), 10)
```
