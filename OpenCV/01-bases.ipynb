````python
import cv2
from matplotlib import pyplot as plt
import numpy as np
````

````python
image = cv2.imread('./images/chihuahua.jpg')
````

````python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("My title")
plt.show()
````

````python
def imshow(img, title=""):
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
````

````python
print(image.shape) # Las imagenes se componen por 3 dimensiones (Red Green Blue)
````

````python
print('Height of Image: {} pixels'.format(int(image.shape[0])))
print('Width of Image: {} pixels'.format(int(image.shape[1])))
print('Depth of Image: {} colors components'.format(int(image.shape[2])))
````

## Grayscale

En gris solo hay una dimension, va de 0 a 255.

Siempre es mejor convertir todas las imagenes a gris antes de procesar, para tener un mucho menor costo comuntacional

````python
# Imagen con escala de grises

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

imshow(gray_image, "in grayscale")
gray_image.shape # una sola dimension
````

## Colors

Las imagenes estan compuestas por 3 dimensiones (red, green & blue)

````python
B, G, R = cv2.split(image)
print(B.shape)
print(G.shape)
print(R.shape)
````

````python
# Each color space on it's on will look like a grayscale as it lacks the other color channels
imshow( R)
````

````python
# Let's create a matrix of zeros 
# image.shape[:2] -> with dimensions of the image h x w  
zeros = np.zeros(image.shape[:2], dtype = "uint8")

imshow(cv2.merge([zeros, zeros, R]))
imshow( cv2.merge([zeros, G, zeros]))
imshow(cv2.merge([B, zeros, zeros]))
````

````python
# Let's re-make the original image, 
merged = cv2.merge([B, G, R]) 
imshow(merged) 
````

````python
merged = cv2.merge([B+100, G, R])imshow(merged)
````
